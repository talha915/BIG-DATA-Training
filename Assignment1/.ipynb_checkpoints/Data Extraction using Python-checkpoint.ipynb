```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/History_of_the_FIFA_World_Cup'
```

```python

table_classes = {"class": ["wikitable", "sortable", "jquery-tablesorter"]}
```

```python
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
fifa_table = soup.findAll('table', table_classes)
```

```python
df = pd.read_html(str(fifa_table))[1]
```

```python
df
```

```python
df.to_csv('FIFA_WC_HISTORY_SINCE_1930.csv')
```

```
import pyspark.sql.functions as F
```

```
orders_df = (spark.read.format("csv") \
             .option("inferSchema", "true") \
             .option("header", "true") \
             .load("dbfs:/FileStore/tables/trainings/trendytech/datasets/week12/orders_201025_223502.csv")
            )
```

```
orders_df.display()
```

```
closed_orders_status = (orders_df \
                        .filter(F.col("order_status") == "CLOSED")
                       )
```

```
pending_payments = (orders_df \
                   .filter(F.col("order_status") == "PENDING_PAYMENT")
                   )
```

```
complete_closed_status = (orders_df \
                         .filter((F.col("order_status") == "CLOSED") | \
                                 (F.col("order_status") == "COMPLETE"))
                         )
```

```
orders_df.filter(F.col("order_customer_id").isNull()).display()
```

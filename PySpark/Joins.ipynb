```
import pyspark.sql.functions as F
```

```
orders_df = (spark.read.format("csv") \
             .option("inferSchema", "true") \
             .option("header", "true") \
             .load("dbfs:/FileStore/tables/trainings/trendytech/datasets/week12/orders_201025_223502.csv")
            )
```

```
customers_df = (spark.read.format("csv") \
               .option("inferSchema", "true") \
                .option("header", "true") \
               .load("dbfs:/FileStore/tables/trainings/trendytech/datasets/week12/customers_201025_223502__1_.csv")
               )
```

###Inner Join

```
order_customer_id = (orders_df \
                     .join(customers_df, \
                          (orders_df.order_customer_id == customers_df.customer_id), \
                          how="inner") 
                    )
```

```
order_customer_id.show(5)
```

```
order_customer_id.count()
```

```
orders_df.count()
```

###Left Join

```
order_customer_id = (orders_df \
                     .join(customers_df, \
                          (orders_df.order_customer_id == customers_df.customer_id), \
                          how="left") 
                    )
```

```
order_customer_id.count()
```

###Right Join

```
order_customer_id = (orders_df \
                     .join(customers_df, \
                          (orders_df.order_customer_id == customers_df.customer_id), \
                          how="right") 
                    )
```

```
order_customer_id.count()
```

###Left Semi Join

```
order_customer_id = (orders_df \
                     .join(customers_df, \
                          (orders_df.order_customer_id == customers_df.customer_id), \
                          how="leftsemi") 
                    )
order_customer_id.count()
```

###Outer Join

```
order_customer_id = (orders_df \
                     .join(customers_df, \
                          (orders_df.order_customer_id == customers_df.customer_id), \
                          how="outer") 
                    )
order_customer_id.count()
```

###Left Anti

```
order_customer_id = (orders_df \
                     .join(customers_df, \
                          (orders_df.order_customer_id == customers_df.customer_id), \
                          how="leftanti") 
                    )
order_customer_id.count()
```

###Full Outer Join

```
order_customer_id = (orders_df \
                     .join(customers_df, \
                          (orders_df.order_customer_id == customers_df.customer_id), \
                          how="full") 
                    )
order_customer_id.count()
```

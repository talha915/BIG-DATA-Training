```
import pyspark.sql.functions as F
```

```
orders_df = (spark.read.format("csv") \
             .option("inferSchema", "true") \
             .option("header", "true") \
             .load("dbfs:/FileStore/tables/trainings/trendytech/datasets/week12/orders_201025_223502.csv")
            )
```

```
orders_df.show(5, truncate=True)
```

##Adding static column

```
new_df = orders_df.withColumn("country", F.lit("USA"))
new_df.show(5)
```

###Renaming Column Name

```
new_df = new_df.withColumnRenamed("country", "country_name")
new_df.show(5)
```

###Drop column

```
new_df = new_df.drop('country_name')
new_df.show(5)
```

```python
import requests
import pandas as pd
import numpy as np
import json
import os
```

```python
url = "https://api.github.com/repos/CSSEGISandData/COVID-19/contents/csse_covid_19_data/csse_covid_19_daily_reports"
```

```python
response = requests.get(url)
```

```python
json_response = response.json()
print(json.dumps(json_response, indent=2))
```

```python
downloaded_urls = []
for i in range(0, len(json_response)):
    if '.csv' in json_response[i]['name']:
        downloaded_urls.append(json_response[i]['download_url'])
```

```python
downloaded_urls[0]
```

```python
relabel = {
    # 'Last Update': 'Last_Update',
    'Country/Region': 'Country_Region',
    'Lat': 'Latitude',
    'Long_': 'Longitude',
    'Province/State': 'Province_State',
}
DF_list= list()
```

```python
def get_dataframe(data, dates):
    for label in data:
        if label in relabel:
            data = data.rename(columns = {label: relabel[label]})
    labels = ['Province_State', 'Country_Region', 'Last_Update', 'Confirmed', 'Deaths', 'Recovered']
#     if 'Last_Update' not in data:
#         data['Last_Update'] = pd.to_datetime(dates)
    for label in labels:
        if label not in data:
            data[label] = np.nan
    return data[labels]
```

```python
for i in range(0, len(downloaded_urls)):
    filename = downloaded_urls[i].split('.')[1]
    dates = filename.split('/')[-1]
    data = pd.read_csv(downloaded_urls[i])
    final_data = get_dataframe(data, dates)
    DF_list.append(final_data)
```

```python
for i in range(0, len(DF_list)):
    if i > 0:
        DF_list[i].to_csv('loaded_data.csv', mode='a', header=False, index=False)
    else:
        DF_list[i].to_csv('loaded_data.csv', index=False)
```

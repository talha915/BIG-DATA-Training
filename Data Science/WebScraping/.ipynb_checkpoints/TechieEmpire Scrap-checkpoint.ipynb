```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url = 'https://www.techieempire.tech/'
req = requests.get(url)
```

```python
soup = BeautifulSoup(req.text, 'html.parser')
print(soup.title)
```

```python
for link in soup.find_all('a'):
    print(link.get('href'))
```

```python
#Extracting Paragraphs
paragraphs = []
headers = []
for para in soup.find_all('p'):
    paragraphs.append(para)
```

```python
headers_type = ['h1', 'h2', 'h3', 'h4', 'h5', 'h6']
for heads in headers_type:
    for headings in soup.find_all(heads):
        headers.append(headings)  
```

```python
print(headers)
```

```python
tags = ['i', 'img', 'style']
all_tag = []
for tag in tags:
    for all_tags in soup.find_all(tag):
        all_tag.append(all_tags)
```

```python
print(all_tag)
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://in.finance.yahoo.com/quotes/recent/view/v1'
```

```python
req = requests.get(url)
soup = BeautifulSoup(req.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
prices = []
def stockPrice():
    price = soup.find('div', {'class': 'Carousel-Mask Pos(r) Ov(h) market-summary M(0) Pos(r) Ov(h) D(ib) Va(t)'}).find_all('span')
    #print(price)
    for i in range(0, len(price), 4):
        prices.append(price[i].text)
        
        #if i['class'] == 'Trsdu(0.3s)', 'Fz(s)', 'Mt(4px)', 'Mb(0px)', 'Fw(b)', 'D(ib)':
        #    print(i)
```

```python
stockPrice()
```

```python
titles = []
for title in soup.find('div', {'class': 'Carousel-Mask Pos(r) Ov(h) market-summary M(0) Pos(r) Ov(h) D(ib) Va(t)'}).find_all('a'):
    titles.append(title['title'])
```

```python
print(prices)
```

```python
import csv
file = open('stock.csv', 'w', newline ='') 
with file:   
    header = ['Title', 'Prices'] 
    writer = csv.DictWriter(file, fieldnames = header) 
      
    # writing data row-wise into the csv file 
    writer.writeheader()
    for i in range(0, len(prices)):
        writer.writerow({'Title' : titles[i],  
                     'Prices': prices[i]}) 
```

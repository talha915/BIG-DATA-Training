# make a lineplot graph between sin and cosine wave with time on x axis

```python
x = np.arange(0,3.2*np.pi,0.1)
y = np.sin(x)
z = np.cos(x)
```

```python
plt.plot(x, y, x, z)
plt.xlabel('Time')
plt.ylabel('amplitude')
plt.show()
```

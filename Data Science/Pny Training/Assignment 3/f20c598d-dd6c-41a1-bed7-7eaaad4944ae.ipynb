# Counting the Most Frequent words in Tags
This section is reading the 5 most frequent words in tags column from dataset and making a list of them. 

```python
list1 = []
for x in tag_data:
  list1.append(x)
TagString = ' '.join(list1)
TagString
```

```python
from collections import Counter 
split_words = TagString.split()
Counter = Counter(split_words)  
most_occur_words = Counter.most_common(5)
  
print(most_occur_words) 
```

# Graph
Plotting the graph for 5 most frequently occuring words in Tags


```python


data = most_occur_words

n_groups = len(data)

words_T = [x[1] for x in data]
Frequency_T = [x[0] for x in data]

fig, ax = plt.subplots()

index = np.arange(n_groups)
bar_width = 0.25

opacity = 0.4

rects1 = plt.bar(index, words_T, bar_width,
                 alpha=opacity,
                 color='b',
                 label='Ocurrences')


plt.xlabel(' Tags ')
plt.ylabel('Frequency')
plt.title('Graph of 5 Most Frequent Tags')
plt.xticks(index + bar_width, Frequency_T)
plt.legend()

plt.tight_layout()
plt.show()
```
